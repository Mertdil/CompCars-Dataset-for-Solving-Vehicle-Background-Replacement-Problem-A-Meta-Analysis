```python
####resim name'leri otamatikleştirmesi make,model,year ve name üstünden gruplama
import os
from openpyxl import Workbook

# Excel dosyasını oluştur
workbook = Workbook()
worksheet = workbook.active

# Excel dosyasına sütun başlıklarını ekle
worksheet.append(["Make ID", "Model ID", "Released Year", "Image Name"])

# Klasör yapısındaki tüm dosyalara eriş
for make in os.listdir('/Users/mertdil/Desktop/data/image'):
    make_path = os.path.join('/Users/mertdil/Desktop/data/image', make)
    if os.path.isdir(make_path):
        for model in os.listdir(make_path):
            model_path = os.path.join(make_path, model)
            if os.path.isdir(model_path):
                for year in os.listdir(model_path):
                    year_path = os.path.join(model_path, year)
                    if os.path.isdir(year_path):
                        for image in os.listdir(year_path):
                            # Resim dosyasının ismini belirle
                            image_name = image.split('.')[0]
                            # Excel dosyasına resim bilgilerini ekle
                            worksheet.append([make, model, year, image_name])

# Excel dosyasını kaydet
workbook.save('/Users/mertdil/Desktop/data/resim_isimleri.xlsx')
```

```python
####resim name'leri otamatikleştirmesi make,model,year,annotation ve name üstünden gruplama

import os
import pandas as pd

# Excel dosyası için boş bir DataFrame oluştur
df = pd.DataFrame(columns=["make_id", "model_id", "released_year", "image_name", "view", "bbox_number", "bbox_cor"])

# Ana klasöre erişim sağla
root_dir = "/Users/mertdil/Desktop/data/label"

# Ana klasör altındaki tüm klasörleri sırayla ziyaret et
for make_id in os.listdir(root_dir):
    make_dir = os.path.join(root_dir, make_id)
    if not os.path.isdir(make_dir):
        continue
    
    for model_id in os.listdir(make_dir):
        model_dir = os.path.join(make_dir, model_id)
        if not os.path.isdir(model_dir):
            continue
        
        for released_year in os.listdir(model_dir):
            year_dir = os.path.join(model_dir, released_year)
            if not os.path.isdir(year_dir):
                continue
            
            for text_file in os.listdir(year_dir):
                if not text_file.endswith(".txt"):
                    continue
                
                # Image name'i almak için dosya adından ".txt" uzantısını kaldır
                image_name = os.path.splitext(text_file)[0]
                
                # Text dosyasını aç ve verileri al
                text_path = os.path.join(year_dir, text_file)
                with open(text_path, "r") as f:
                    lines = f.readlines()
                    # Verileri DataFrame'e ekle
                    for i in range(0, len(lines), 3):
                        view = lines[i].strip()
                        bbox_number = lines[i+1].strip()
                        bbox_cor = lines[i+2].strip()
                        df = df.append({
                            "make_id": make_id,
                            "model_id": model_id,
                            "released_year": released_year,
                            "image_name": image_name,
                            "view": view,
                            "bbox_number": bbox_number,
                            "bbox_cor": bbox_cor
                        }, ignore_index=True)

# DataFrame'i Excel dosyasına yazdır
df.to_excel("/Users/mertdil/Desktop/data/annotation.xlsx", index=False)
```
